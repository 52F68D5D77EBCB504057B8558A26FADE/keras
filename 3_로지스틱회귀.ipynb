```python
%config InlineBackend.figure_format = "retina"
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1,2,3,4,5,6,7,8]) # 종양 크기
y = np.array([0,0,0,0,1,1,1,1]) # 악성 유무

plt.scatter(x, y, edgecolor="black")
plt.xlabel("tumor size")
plt.ylabel("cancer")
plt.show()
```

```python
# 선형 회귀 분석을 사용하여 암을 분류해 봅니다.
# 바이어스에 대해서도 편미분을 수행하여 업데이트합니다.
import numpy as np

def mse(y_hat, y):
    return ((y_hat - y) ** 2).mean()

# 1. 데이터 준비
x = np.array([1,2,3,4,5,6,7,8]) # 종양 크기
y = np.array([0,0,0,0,1,1,1,1]) # 악성 유무

# 2. 가중치와 바이어스를 초기화 합니다.
w = 0
b = 0

# 3. 학습을 수행합니다.
step = 0.01
n_epochs = 3000
for epoch in range(n_epochs+1):
    y_hat = w * x + b

    w_grad = (2 / len(x)) * np.sum((y_hat - y) * x)
    b_grad = (2 / len(x)) * np.sum((y_hat - y) * 1)

    w -= step * w_grad
    b -= step * b_grad
    
    if (epoch % 100) == 0:
        err = mse(y_hat, y)
        print(f"epoch: {epoch:4d}/{n_epochs}, err: {err:f}")
```

```python
# 분류를 잘 수행했는지 확인합니다.
for i, value in enumerate(y_hat):
    if value >= 0.5: # 임계점(threshold)을 0.5로 하여 분류를 수행합니다.
        print(f"{i}: 악성")
    else:
        print(f"{i}: 양성")
```

```python
# 그래프를 잘 살펴보면 기울기가 낮아져서 기존의 악성 종양이 양성으로 분류된 것을 확인할 수 있다.
# 때문에 회귀 직선은 이진 분류에 사용하기는 어렵다.
# 새로운 데이터가 입력된다 하더라도 기울기의 변화가 없어야 한다.
# 그리고 모든 예측 값에 대하여 1또는 0으로 수렴해야 한다.
# 이는 악성의 유무를 확률로 판단해야 하기 때문이다. 
# 이러한 특성을 만족하는 함수가 존재하는데 바로 시그모이드 함수이다.
# 이 함수는 모든 입력에 대하여 0또는 1로 수렴한다.
# 시그모이드 함수
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10,10,100)
y = sigmoid(x)
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("sigmoid(x)")
plt.show()
```

```python
# 다음은 시그모이드 함수가 도입된 코드이다.
import numpy as np

# binary cross entropy
def bce(y_hat, y):
    return np.mean(-y*np.log(y_hat) - (1-y)*np.log(1-y_hat))

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x = np.array([0,1,2,3,4,5,6,7,15])
y = np.array([0,0,0,0,1,1,1,1,1])

w = 0
b = 0

lr = 0.1
n_epochs = 1000
for epoch in range(n_epochs+1):
    y_hat = sigmoid(w*x + b)

    if (epoch % 100) == 0:
        err = bce(y_hat, y)
        print(f"epoch: {epoch:5d}, err: {err}")

    w_diff = (2 / len(x)) * np.sum((y_hat - y) * x)
    b_diff = (2 / len(x)) * np.sum((y_hat - y) * 1)

    w -= lr * w_diff
    b -= lr * b_diff
```

```python
for i, value in enumerate(y_hat):
    if value >= 0.5:
        print(f"{i}: 악성")
    else:
        print(f"{i}: 양성")
```

```python
# 케라스의 사용
# 1. 텐서의 사용
import tensorflow as tf

# 텐서플로는 sigmoid와 bce 함수를 제공합니다.
sigmoid = tf.keras.activations.sigmoid
bce = tf.keras.losses.binary_crossentropy

x = tf.constant([0,1,2,3,4,5,6,7,15], dtype=tf.float32)
y = tf.constant([0,0,0,0,1,1,1,1,1], dtype=tf.float32)

w = tf.Variable(1.0)
b = tf.Variable(1.0)

lr = 0.1
epochs = 20000
for epoch in range(epochs+1):
    z = w * x  + b
    y_hat = sigmoid(z)

    w_grad = 2 * tf.reduce_mean((y_hat - y) * x)
    b_grad = 2 * tf.reduce_mean((y_hat - y) * 1)

    w.assign_sub(lr * w_grad)
    b.assign_sub(lr * b_grad)

    if (epoch % 1000) == 0:
        cost = bce(y, y_hat)
        print(f"epoch: {epoch:>5}/{epochs}, err: {cost:.10f}")
```

```python
pred = tf.cast(y_hat >= 0.5, dtype=tf.float32)
acc = tf.reduce_mean(tf.cast(pred == y, dtype=tf.float32))
print(acc.numpy())
```

```python
# 2. 자동 미분의 도입
# 자동 미분의 도입
import tensorflow as tf

# 텐서플로는 sigmoid와 bce 함수를 제공합니다.
sigmoid = tf.keras.activations.sigmoid
bce = tf.keras.losses.binary_crossentropy

x = tf.constant([0,1,2,3,4,5,6,7,15], dtype=tf.float32)
y = tf.constant([0,0,0,0,1,1,1,1,1], dtype=tf.float32)

w = tf.Variable(1.0)
b = tf.Variable(1.0)

lr = 0.1
epochs = 20000
for epoch in range(epochs+1):
    with tf.GradientTape() as tape:
        y_hat = sigmoid(w*x  + b)
        cost = bce(y, y_hat)

    # w_grad = 2 * tf.reduce_mean((y_hat - y) * x)
    # b_grad = 2 * tf.reduce_mean((y_hat - y) * 1)
    w_grad, b_grad = tape.gradient(cost, [w, b])

    w.assign_sub(lr * w_grad)
    b.assign_sub(lr * b_grad)

    if (epoch % 1000) == 0:
        cost = bce(y, y_hat)
        print(f"epoch: {epoch:>5}/{epochs}, err: {cost:.10f}")
```

```python
pred = tf.cast(y_hat >= 0.5, dtype=tf.float32)
acc = tf.reduce_mean(tf.cast(pred == y, dtype=tf.float32))
print(acc.numpy())
```
