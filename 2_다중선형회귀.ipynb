```python
# 케라스의 사용
# 1. 텐서의 도입

import numpy as np
import tensorflow as tf

def mse(y_hat, y):
    return tf.reduce_mean(tf.square(y_hat - y))

x1 = tf.constant([73, 93, 89, 96, 73], dtype=tf.float32)
x2 = tf.constant([80, 88, 91, 98, 66], dtype=tf.float32)
x3 = tf.constant([75, 93, 90, 100, 70], dtype=tf.float32)
y = tf.constant([152, 185, 180, 196, 142], dtype=tf.float32)

w1 = tf.Variable(1.0)
w2 = tf.Variable(1.0)
w3 = tf.Variable(1.0)
b = tf.Variable(1.0)

lr = 0.00001  # 주의! 학습률이 높으면 발산할 수 있습니다.
epochs = 30000
for epoch in range(epochs+1):
    y_hat = w1*x1 + w2*x2 + w3*x3 + b

    w1_grad = 2 * tf.reduce_mean((y_hat-y) * x1)
    w2_grad = 2 * tf.reduce_mean((y_hat-y) * x2)
    w3_grad = 2 * tf.reduce_mean((y_hat-y) * x3)
    b_grad = 2 * tf.reduce_mean((y_hat-y) * 1)

    w1.assign_sub(lr * w1_grad)
    w2.assign_sub(lr * w2_grad)
    w3.assign_sub(lr * w3_grad)
    b.assign_sub(lr * b_grad)

    if (epoch % 1000) == 0:
        cost = mse(y_hat, y)
        print(f"epoch: {epoch:>5}/{epochs}, err: {cost:.10f}")
```

```python
# 2. 행렬의 도입
import numpy as np
import tensorflow as tf

def mse(y_hat, y):
    return tf.reduce_mean(tf.square(y_hat - y))

x = tf.constant([[73, 80, 75], 
                 [93, 88, 93], 
                 [89, 91, 80], 
                 [96, 98, 99],   
                 [73, 66, 70]], dtype=tf.float32)  
y = tf.constant([152, 185, 180, 196, 142], dtype=tf.float32)
y = tf.reshape(y, (-1, 1))

# 입력의 특성 수: 3
# 출력의 특성 수: 1
w = tf.Variable(tf.ones((3,1)))
b = tf.Variable(1.0)

lr = 0.00001  # 주의! 학습률이 높으면 발산할 수 있습니다.
epochs = 30000
for epoch in range(epochs+1):
    y_hat = x @ w + b # y_hat = tf.matmul(x, w) + b 

    err = y_hat - y                                  #  x.T     err
    w_grad = (2 / len(x)) * (tf.transpose(x) @ err)  # (3,5) @ (5,1) = (3,1)
    b_grad = (2 / len(x)) * tf.reduce_sum(1 * err)

    w.assign_sub(lr * w_grad)
    b.assign_sub(lr * b_grad)

    if (epoch % 1000) == 0:
        cost = mse(y_hat, y)
        print(f"epoch: {epoch:>5}/{epochs}, err: {cost:.10f}")

```

```python
# 3. 자동 미분의 도입
import numpy as np
import tensorflow as tf

def mse(y_hat, y):
    return tf.reduce_mean(tf.square(y_hat - y))

x = tf.constant([[73, 80, 75], 
                 [93, 88, 93], 
                 [89, 91, 80], 
                 [96, 98, 99],   
                 [73, 66, 70]], dtype=tf.float32)  
y = tf.constant([152, 185, 180, 196, 142], dtype=tf.float32)
y = tf.reshape(y, (-1, 1))

# 입력의 특성 수: 3
# 출력의 특석 수: 1
w = tf.Variable(tf.ones((3,1)))
b = tf.Variable(1.0)

lr = 0.00001  # 주의! 학습률이 높으면 발산할 수 있습니다.
epochs = 30000
for epoch in range(epochs+1):
    with tf.GradientTape() as tape:
        y_hat = x @ w + b # y_hat = tf.matmul(x, w) + b
        cost = mse(y_hat, y)

    w_grad, b_grad = tape.gradient(cost, [w, b])

    w.assign_sub(lr * w_grad)
    b.assign_sub(lr * b_grad)

    if (epoch % 1000) == 0:
        print(f"epoch: {epoch:>5}/{epochs}, err: {cost:.10f}")
```

```python
# 4. 옵티마이저의 도입
import numpy as np
import tensorflow as tf

def mse(y_hat, y):
    return tf.reduce_mean(tf.square(y_hat - y))

x = tf.constant([[73, 80, 75], 
                 [93, 88, 93], 
                 [89, 91, 80], 
                 [96, 98, 99],   
                 [73, 66, 70]], dtype=tf.float32)  
y = tf.constant([152, 185, 180, 196, 142], dtype=tf.float32)
y = tf.reshape(y, (-1, 1))

# 입력의 특성 수: 3
# 출력의 특석 수: 1
w = tf.Variable(tf.ones((3,1)))
b = tf.Variable(1.0)

optimizer = tf.optimizers.SGD(learning_rate=0.00001)  # lr = 0.00001
epochs = 30000
for epoch in range(epochs+1):
    with tf.GradientTape() as tape:
        y_hat = x @ w + b # y_hat = tf.matmul(x, w) + b
        cost = mse(y_hat, y)
  
    w_grad, b_grad = tape.gradient(cost, [w, b])

    # w.assign_sub(lr * w_grad)
    # b.assign_sub(lr * b_grad)
    optimizer.apply_gradients(zip([w_grad, b_grad], [w, b]))

    if (epoch % 1000) == 0:
        print(f"epoch: {epoch:>5}/{epochs}, err: {cost:.10f}")
```

```python
# 5. 케라스의 도입
import numpy as np
import tensorflow as tf

x = tf.constant([[73, 80, 75], 
                 [93, 88, 93], 
                 [89, 91, 80], 
                 [96, 98, 99],   
                 [73, 66, 70]], dtype=tf.float32)  
y = tf.constant([152, 185, 180, 196, 142], dtype=tf.float32)
y = tf.reshape(y, (-1, 1))

# 입력과 출력의 형태를 정의합니다.
input_ = tf.keras.layers.Input(shape=(3,)) # 독립 변수의 특성(컬럼)이 3개입니다.
output_ = tf.keras.layers.Dense(1)(input_)  # 종속 변수의 특성(컬럼)이 1개입니다.

# 모델을 생성합니다.
model = tf.keras.models.Model(input_, output_)
model.compile(optimizer=tf.keras.optimizers.SGD(learning_rate=0.00001), loss="mse")

# 주어진 데이터를 사용하여 학습을 수행합니다.
model.fit(x, y, epochs=5000)
```
